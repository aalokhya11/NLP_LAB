1. Implement Word Embeddings on following text)
    words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england']. [CO1]

```python
import numpy as np
from gensim.models import KeyedVectors
import gensim.downloader as api

# Step 1: Load the pre-trained GloVe model (100-dimensional)
glove_model = api.load("glove-wiki-gigaword-100")  # 100-dimensional vectors

# Step 2: List of words
words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england']

# Step 3: Extract word embeddings for each word
word_embeddings = np.array([glove_model[word] for word in words])

# Step 4: Display the word embeddings
for word, embedding in zip(words, word_embeddings):
    print(f"Word: {word}")
    print(f"Embedding: {embedding}\n")
```

2.Implement Word Embeddings using PCA (Principal Component Analysis) on following text)
    words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england']. [CO1]

```python
import numpy as np
from gensim.models import KeyedVectors
import gensim.downloader as api
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Step 1: Load the pre-trained GloVe model (100-dimensional)
glove_model = api.load("glove-wiki-gigaword-100")  # 100-dimensional vectors

# Step 2: List of words
words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england']

# Step 3: Extract word embeddings for each word
word_embeddings = np.array([glove_model[word] for word in words])

# Step 4: Apply PCA to reduce dimensionality to 2D
pca = PCA(n_components=2)
word_embeddings_2d = pca.fit_transform(word_embeddings)

# Step 5: Plotting the 2D word embeddings
plt.figure(figsize=(8, 6))
for i, word in enumerate(words):
    plt.scatter(word_embeddings_2d[i, 0], word_embeddings_2d[i, 1])
    plt.text(word_embeddings_2d[i, 0] + 0.01, word_embeddings_2d[i, 1] + 0.01, word, fontsize=12)
plt.title("Word Embeddings Visualized using PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()
```

3.Execute Dependency Parsing on any text or on following text [CO1]
The quick brown fox jumps over the lazy dog

```python
import spacy
from spacy import displacy

# Step 1: Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Step 2: Input text for dependency parsing
text = "The quick brown fox jumps over the lazy dog."

# Step 3: Process the text with spaCy to obtain the dependency parse
doc = nlp(text)

# Step 4: Display dependency parsing result
for token in doc:
    print(f"Word: {token.text}, POS: {token.pos_}, Dependency: {token.dep_}, Head: {token.head.text}")

# Step 5: Visualize the dependency parse
displacy.render(doc, style="dep", jupyter=True)
```
